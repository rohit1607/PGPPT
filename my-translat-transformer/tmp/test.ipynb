```python
import numpy as np
import matplotlib.pyplot as plt

def rasterize_polygon(vertices, resolution):
    # Find bounding box
    min_x = min(vertices, key=lambda vertex: vertex[0])[0]
    max_x = max(vertices, key=lambda vertex: vertex[0])[0]
    min_y = min(vertices, key=lambda vertex: vertex[1])[1]
    max_y = max(vertices, key=lambda vertex: vertex[1])[1]

    # Create grid
    grid = np.zeros((resolution, resolution))

    # Scanline Fill Algorithm
    for y in range(min_y, max_y + 1):
        intersections = []
        for i in range(len(vertices)):
            current = vertices[i]
            next = vertices[(i + 1) % len(vertices)]
            if current[1] <= y < next[1] or next[1] <= y < current[1]:
                x_intersection = int(current[0] + (float(y - current[1]) / (next[1] - current[1])) * (next[0] - current[0]))
                intersections.append(x_intersection)

        intersections.sort()
        for i in range(0, len(intersections), 2):
            for x in range(intersections[i], intersections[i + 1] + 1):
                grid[y][x] = 1

    return grid

def mark_partially_filled_cells(grid):
    for y in range(grid.shape[0]):
        row = grid[y]
        first_filled = -1
        for x in range(grid.shape[1]):
            if row[x] == 1:
                if first_filled == -1:
                    first_filled = x
            elif first_filled != -1:
                for px in range(first_filled, x):
                    grid[y][px] = 0.5
                first_filled = -1
        if first_filled != -1:
            for px in range(first_filled, grid.shape[1]):
                grid[y][px] = 0.5

    return grid

# Example usage
vertices = [(2, 2), (8, 2), (8, 8), (5, 10), (2, 8)]
resolution = 100

grid = rasterize_polygon(vertices, resolution)
grid = mark_partially_filled_cells(grid)
v = np.array(vertices)
plt.scatter(v[:,0], v[:,1])

# Display grid
plt.imshow(grid, cmap='gray', origin='lower')
plt.show()
```

```python
import math
import random

def generate_random_points_on_circle(n, radius, center=(0, 0)):
    points = []
    for _ in range(n):
        angle = random.uniform(0, 2*math.pi)
        x = center[0] + radius * math.cos(angle)
        y = center[1] + radius * math.sin(angle)
        points.append((x, y))
    return points

# Example usage
num_points = 5
circle_radius = 10
circle_center = (0, 0)

random_points = generate_random_points_on_circle(num_points, circle_radius, circle_center)
print(random_points)
```
